## Qué hay acá?

Tomando los datos de 1_intermediate-vectores y reduciendo la dimensionalidad de los vectores de palabras con PCA. Genero vectores en 3 dimensiones para poder visualizarlos en un gráfico 3D.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../data/1_intermediate/1_intermediate-vectores.json', orient='records', lines=True,)
print(df.shape)
df.head()
```

```python
# convert rows of df in an array
df['vectores'] = df.apply(lambda x: x.iloc[0:].values, axis=1)
df.head()
```

```python
df.iloc[0]['vectores'].shape
```

```python
# convert df['vectores'] in numpy arrays
df['vectores'] = df['vectores'].apply(lambda x: np.array(x))
df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
vectores = df['vectores'].values.tolist()
vectores = np.array(vectores)
vectores.shape
```

```python
pca = PCA(n_components=3)
pca.fit(vectores)
```

```python
vectores_transformed = pca.transform(vectores)
vectores_transformed.shape
```

```python
vectores_transformed[0]
```

```python
# create a dataframe with the transformed vectors
df_transformed = pd.DataFrame(vectores_transformed, columns=['x', 'y', 'z'])
df_transformed.head()
```

```python
# export to json
df_transformed.to_json('../data/1_intermediate/1_intermediate-vectores-reducidos-pca.json', orient='records', lines=True)
```
